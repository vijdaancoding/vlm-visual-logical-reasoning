### Creating the Chat Template to fit Dataset

```python
system_message = """
You are a highly advanced Vision Language Model (VLM), specialized in analyzing, describing, and interpreting visual data.
Your task is to process and extract meaningful insights from images, videos, and visual patterns, leveraging multimodal understanding to provide accurate and contextually relevant information.

When you respond:
– Wrap your final answer in <answer>…</answer>.

Example:
<answer>
Your succinct final answer here.
</answer>
"""
```

```python
def cot_convert(sample, system_message):
    messages = []
    if "system" in sample:  # Checking for sys msgs in dataset
        messages.append({
            "role": "system",
            "content": [{"type": "text", "text": sample["system"]}],
        })
    else:
        # using our sys msg 
        messages.append({
            "role:": "system",
            "content": [{"type": "text", "text": system_message}],        
        })
    # extracting each element 
    problem = sample.get("problem")
    solution = sample.get("solution")
    image = sample.get("image")

    # User side contains the problem + img
    messages.append({
        "role": "user",
        "content": [
            {"type": "text", "text": problem},
            {"type": "image", "image": image},
        ]
    })
    messages.append({
        "roles": "assistant",
        "content": solution, #removed thinking and answer is only solution
    })

    sample["messages"] = messages
    return sample    
```

### Loading Model

We use bfloat16. Why? Because bfloat can represent greater range of values than float16. However, its less precise than float16. But when it comes to training large models bfloat16 is more stable

```python
if device == "cuda":
    bnb_config =BitsAndBytesConfig(
        load_in_4bit=True,  # 4 bit precision 
        bnb_4bit_use_double_quant=True, # double quant - first to 8bit then 4bit - this maintains accuracy
        bnb_4bit_quant_type="nf4", # Normal Float 4 - better than int4 for representing values closer to 0. 
        bnb_4bit_compute_dtype=torch.bfloat16 # Perform forward/backward in bfloat16
    )
    model = Qwen2VLForConditionalGeneration.from_pretrained(
        MODEL_ID,
        device_map="auto", # Place layers in GPU
        quantization_config=bnb_config, 
        use_cache=False # During training we dont save past weights
    )
else:
    model = Qwen2VLForConditionalGeneration.from_pretrained(
        MODEL_ID,
        use_cache=False
    ) 

processor = Qwen2VLProcessor.from_pretrained(MODEL_ID)
processor.tokenizer.padding_side = "right"
```

### Data Analysis

```python
def calculate_token_lengths(train_dataset, processor, max_seq_len=2048):
  
    token_lengths = []
    exceeding_examples = []
    
    for idx, example in enumerate(train_dataset):
        try:
            # Convert the example to messages format using convert_example
            messages = example["messages"]
            
            # Apply chat template as in collate_fn
            text = processor.apply_chat_template(
                messages, tokenize=False, add_generation_prompt=True
            )
            
            # Get images using process_vision_info
            imgs, _ = process_vision_info(messages)
            
            # Tokenize exactly as in collate_fn
            inputs = processor(
                text=[text],
                images=imgs,
                return_tensors="pt"
            )
            
            # Get token length
            input_ids = inputs["input_ids"][0]
            token_length = len(input_ids)
            token_lengths.append(token_length)
            
            # Track examples exceeding threshold
            if token_length > max_seq_len:
                exceeding_examples.append((idx, token_length))
                
        except Exception as e:
            print(f"Skipping example {idx} due to error: {e}")
            continue
            
    return token_lengths, exceeding_examples

# Run the calculation
token_lengths, exceeding_examples = calculate_token_lengths(train_dataset, processor, max_seq_len=2048)
```

```python
print(f"Total examples processed: {len(token_lengths)}")
print(f"Max tokens: {max(token_lengths)}")
print(f"Min tokens: {min(token_lengths)}")
print(f"Average tokens: {sum(token_lengths) / len(token_lengths):.2f}")
print(f"95th percentile: {sorted(token_lengths)[int(0.95 * len(token_lengths))]}")
print(f"Examples exceeding {2048} tokens: {len(exceeding_examples)} ({len(exceeding_examples)/len(token_lengths):.2%})")

# Plot distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(token_lengths, bins=50, color='skyblue')
plt.title("Token Length Distribution")
plt.xlabel("Number of tokens")
plt.ylabel("Number of examples")
plt.axvline(x=2048, color='red', linestyle='--', label='MAX_SEQ_LEN = 2048')
plt.legend()
plt.grid(True)
plt.show()
```

```python
if len(token_lengths) > 0:
    try:
        # Check if we can analyze by prompts/inputs
        problem_lengths = []
        solution_lengths = []
        images_count = []
        
        for idx, example in enumerate(train_dataset):
            if idx >= len(token_lengths):
                continue
                
            # Get problem length (text characters)
            problem = example.get("problem", "")
            problem_lengths.append(len(problem) if isinstance(problem, str) else 0)
            
            # Get solution length (text characters) 
            solution = example.get("solution", "")
            solution_lengths.append(len(solution) if isinstance(solution, str) else 0)
            
            # Count images
            messages = example.get("messages", [])
            img_count = 0
            for msg in messages:
                if msg.get("role") == "user":
                    for item in msg.get("content", []):
                        if isinstance(item, dict) and item.get("type") == "image":
                            img_count += 1
            images_count.append(img_count)
        
        # Calculate correlations
        import numpy as np
        problem_corr = np.corrcoef(problem_lengths, token_lengths)[0, 1]
        solution_corr = np.corrcoef(solution_lengths, token_lengths)[0, 1]
        
        print("\nCorrelation Analysis:")
        print(f"Correlation between problem length and token count: {problem_corr:.3f}")
        print(f"Correlation between solution length and token count: {solution_corr:.3f}")
        
        # Group by image count
        image_groups = {}
        for i, count in enumerate(images_count):
            if count not in image_groups:
                image_groups[count] = []
            image_groups[count].append(token_lengths[i])
        
        print("\nAverage token count by number of images:")
        for img_count in sorted(image_groups.keys()):
            avg_tokens = sum(image_groups[img_count]) / len(image_groups[img_count])
            print(f"{img_count} image(s): {avg_tokens:.1f} tokens (n={len(image_groups[img_count])})")
    
    except Exception as e:
        print(f"Could not perform additional analysis: {e}")
```

### LoRA

* **alpha**: Is used as a scaling factor to make sure the training is stable. What does that mean? If Alpha is too large then (A * B) * (very large num) may result in a LoRA weight matrix that is significantly more over powered than the orignial weight matrix - could also lead to gradient explosions. If Alpha is too small it means that the LoRA weight matrix is too weak as compared to the orginal.
* **r**: Matrix A (r, d) * Matrix (d, r) * (alpha / r) = W (d , d)
* **bias**: We chose none but you can choose to train biases -> it might improve performance
* **task_type**: CAUSAL_LM is needed since we need Auto Regressive Loss for text generation. CAUSAL_LM is future masking so:
Actual: The cat sat on the ___ Predicted: mat

```python
peft_config = LoraConfig(
    lora_alpha=16, # (A * B) * (alpha / r)
    lora_dropout=0.1,
    # Small r -> less mem but less performance
    r=8, # Matrix A (r, d) * Matrix B (d, r) = Weight Matrix (d, d)
    bias="none",
    # so basically mostly focusing on the attention layers
    target_modules=["q_proj", "v_proj"], # Only training the query and key projections
    task_type="CAUSAL_LM"
)

print(f"Before Adapter Parameters: {model.num_parameters()}")
peft_model = get_peft_model(model, peft_config)
peft_model.print_trainable_parameters()
```

```python
# no way I explaining this
NUM_CPU_CORES = 4
training_args = SFTConfig(
    output_dir="./output",
    # dataloader_num_workers=NUM_CPU_CORES,
    # dataloader_pin_memory=True,
    fp16=True,
    num_train_epochs=EPOCHS,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
    gradient_checkpointing=GRADIENT_CHECKPOINTING,
    learning_rate=LEARNING_RATE,
    logging_steps=LOGGING_STEPS,
    eval_steps=EVAL_STEPS,
    eval_strategy=EVAL_STRATEGY,
    save_strategy=SAVE_STRATEGY,
    save_steps=SAVE_STEPS,
    metric_for_best_model=METRIC_FOR_BEST_MODEL,
    load_best_model_at_end=LOAD_BEST_MODEL_AT_END,
    max_grad_norm=MAX_GRAD_NORM,
    warmup_steps=WARMUP_STEPS,
    dataset_kwargs=DATASET_KWARGS,
    max_seq_length=MAX_SEQ_LEN,
    remove_unused_columns = REMOVE_UNUSED_COLUMNS,
    optim=OPTIM,
)
```

```python
def collate_fn(examples):
    texts = [
        processor.apply_chat_template(example["messages"], tokenize=False, add_generation_prompt=True)
        for example in examples
    ]
    image_inputs = []
    for example in examples:
        imgs, vids = process_vision_info(example["messages"])
        image_inputs.append(imgs)
    batch = processor(
        text=texts,
        images=image_inputs,
        return_tensors="pt",
        padding=True,
        truncation=True
    )
    labels = batch["input_ids"].clone()
    labels[labels == processor.tokenizer.pad_token_id] = -100 # To hide pad tokens
    image_token_id = processor.tokenizer.convert_tokens_to_ids(processor.image_token)
    labels[labels == image_token_id] = -100
    batch["labels"] = labels

    return batch
```

### Training

```python
trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=collate_fn,
    peft_config=peft_config,
    processing_class=processor.tokenizer,
)
```

```python
print("-"*30)
print("Initial Evaluation")
metric = trainer.evaluate()
print(metric)
print("-"*30)

print("Training")
trainer.train()
print("-"*30)
```

### Saving Model

### Cleaning Memory

```python
import gc
import time

def clear_memory():
    if "inputs" in globals():
        del globals()["inputs"]
    if "model" in globals():
        del globals()["model"]
    if "processor" in globals():
        del globals()["processor"]
    if "trainer" in globals():
        del globals()["trainer"]
    if "peft_model" in globals():
        del globals()["peft_model"]
    if "bnb_config" in globals():
        del globals()["bnb_config"]
    time.sleep(2)

    gc.collect()
    time.sleep(2)
    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    time.sleep(2)
    gc.collect()
    time.sleep(2)

    print(f"GPU allocated memory: {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
    print(f"GPU reserved memory: {torch.cuda.memory_reserved() / 1024**3:.2f} GB")


clear_memory()
```

### Inferencing

```python
def text_gen(example):
    text = processor.apply_chat_template(example["messages"], tokenize=False, add_generation_prompt=True)
        
    print(f"Prompt: {text}")
    print("-"*30)

    image_inputs = []
    imgs, vids = process_vision_info(example["messages"])
    image_inputs.append(imgs)
    
    inputs = processor(
        text=text,
        images=image_inputs,
        return_tensors="pt"
    )

    inputs = inputs.to(device)

    generated_ids = model.generate(**inputs, max_new_tokens=MAX_SEQ_LEN)

    output_text = processor.batch_decode(
        generated_ids, skip_special_tokens=True
    )
    del inputs

    actual_answer = example["messages"][2]['content']
    return output_text[0], actual_answer
```

### LoRA

* **alpha**: Is used as a scaling factor to make sure the training is stable. What does that mean? If Alpha is too large then (A * B) * (very large num) may result in a LoRA weight matrix that is significantly more over powered than the orignial weight matrix - could also lead to gradient explosions. If Alpha is too small it means that the LoRA weight matrix is too weak as compared to the orginal.
* **r**: Matrix A (r, d) * Matrix (d, r) * (alpha / r) = W (d , d)
* **bias**: We chose none but you can choose to train biases -> it might improve performance
* **task_type**: CAUSAL_LM is needed since we need Auto Regressive Loss for text generation. CAUSAL_LM is future masking so:
Actual: The cat sat on the ___ Predicted: mat

```python
peft_config = LoraConfig(
    lora_alpha=16, # (A * B) * (alpha / r)
    lora_dropout=0.1,
    # Small r -> less mem but less performance
    r=8, # Matrix A (r, d) * Matrix B (d, r) = Weight Matrix (d, d)
    bias="none",
    # so basically mostly focusing on the attention layers
    target_modules=["q_proj", "v_proj"], # Only training the query and key projections
    task_type="CAUSAL_LM"
)

print(f"Before Adapter Parameters: {model.num_parameters()}")
peft_model = get_peft_model(model, peft_config)
peft_model.print_trainable_parameters()
```

### Cleaning Memory

```python
import gc
import time

def clear_memory():
    if "inputs" in globals():
        del globals()["inputs"]
    if "model" in globals():
        del globals()["model"]
    if "processor" in globals():
        del globals()["processor"]
    if "trainer" in globals():
        del globals()["trainer"]
    if "peft_model" in globals():
        del globals()["peft_model"]
    if "bnb_config" in globals():
        del globals()["bnb_config"]
    time.sleep(2)

    gc.collect()
    time.sleep(2)
    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    time.sleep(2)
    gc.collect()
    time.sleep(2)

    print(f"GPU allocated memory: {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
    print(f"GPU reserved memory: {torch.cuda.memory_reserved() / 1024**3:.2f} GB")


clear_memory()
```

### Inferencing

```python
def text_gen(example):
    text = processor.apply_chat_template(example["messages"], tokenize=False, add_generation_prompt=True)
        
    print(f"Prompt: {text}")
    print("-"*30)

    image_inputs = []
    imgs, vids = process_vision_info(example["messages"])
    image_inputs.append(imgs)
    
    inputs = processor(
        text=text,
        images=image_inputs,
        return_tensors="pt"
    )

    inputs = inputs.to(device)

    generated_ids = model.generate(**inputs, max_new_tokens=MAX_SEQ_LEN)

    output_text = processor.batch_decode(
        generated_ids, skip_special_tokens=True
    )
    del inputs

    actual_answer = example["messages"][2]['content']
    return output_text[0], actual_answer
```
